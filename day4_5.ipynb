# 5일차 2.22 (Home Controller part Two)

- mixins 개념에 대해 배움
- mixin은 pug 함수의 일종임
- mixin에는 인자가 하나 필요함
- mixin은 공통적으로 반복되는 html을 편하게 다루게 해줌

```javascript
// views/home.pug

extends layouts/main.pug
include mxins/videoBlock //videoBlock 을 include 해줌

block content
	.videos
		each video in videos
			+videoBlock({
				title:video.title,
				views:video.views,
				videoFile:video.videoFile
			})
```

```javascript
// views/mixins/videoBlock.pug

mixin videoBlock(video = {})
	.videoBlock
		video.videoBlock__thumbnail(src=video.videoFile, controls=true)
		h4.videoBlock__title=video.title
		h6.videoBlock__views=video.views
```

# 5일차 3.4 (Comment Model)

- model에 foregin key(reference) 작성법 배움
- 비디오 model에 댓글 model 연관 시킴

```javascript
// models/Comment.js

import mongoose from "mongoose";

const CommentSchema = new mongoose.Schema({
	text: {
		type: String,
		required: "Text is required"
	},
	createdAt: {
		type: Date,
		default: Date.now
	},
	// 한 댓글에 한 비디오 ID를 줘도 되고,
	// 한 비디오에 여러 댓글(리스트)로 줘도 됨(선택사항)
	// video: {
	// 	type: mongoose.Schema.Types.ObjectId,
	// 	ref: "Video"
	// }
});

const model = mongoose.model("Comment", CommentSchema);
export default model;
```

```javascript
// models/Video.js

import mongoose from "mongoose"

const VideoSchema = new mongoose.Schema({
	fileUrl: {
		type: String,
		required: "File URL is required"
	},
	title: {
		type: String,
		required: "Title is required"
	},
	description: String,
	views: {
		type: Number,
		default: 0
	},
	createdAt: {
		type: Date,
		default: Date.now
	},
    // 한 비디오에 여러 댓글을 가짐
	comments: [{
		type: mongoose.Schema.Types.ObjectId,
		ref: "Comment"
	}]
});

const model = mongoose.model("Video", VideoSchema);
export default model;
```

# 5일차 3.5 (Home Controller Finished)

- controller에서 에러가 나도 페이지 렌더링을 유지할 수 있도록 하는 법 : try ... catch 활용

```javascript
// controllers/videoController.js

import routes from "../routes"
import Video from "../models/Video"

export const home = async (req, res) => { // async ... await
	try {
		const videos = await Video.find({});
		res.render("home", { pageTitle: "Home", videos });
	} catch (error) {
		// 에러가 나도 페이지는 그대로 돌아가도록 할 수 있음
		console.log(error);
		res.render("home", { pageTitle: "Home", videos: [] }); // videos 빈 리스트를 가지게 됨
	}
}
```

# 5일차 3.6 (Uploading and Creating a Video)

- multer 사용법(파일 업로드, 저장된 파일 경로 가져오기, 등)
- mongodb에 객체 저장하는 법(model.create({})

```javascript
// controllers/videoController.js

import routes from "../routes"
import Video from "../models/Video"

export const home = async (req, res) => {
	try {
		const videos = await Video.find({});
		res.render("home", { pageTitle: "Home", videos });
	} catch (error) {
		// 에러가 나도 페이지는 그대로 돌아가도록 할 수 있음
		console.log(error);
		res.render("home", { pageTitle: "Home", videos: [] });
	}
}
export const search = (req, res) => {
	const {
		query: { term: searchingBy }
	} = req;
	res.render("search", { pageTitle: "Search", searchingBy, videos });
}
export const getUpload = (req, res) => {
	res.render("upload", { pageTitle: "Upload" });
}
export const postUpload = async (req, res) => {
	const {
		body: { title, description },
		file: { path }
	} = req;
	const newVideo = await Video.create({ // db에 저장함
		fileUrl: path,
		title,
		description
	});
	console.log(newVideo);
	res.redirect(routes.videoDetail(newVideo.id)); // db 저장 후 해당 video id 로 바로 이동함
}
export const videoDetail = (req, res) => res.render("videoDetail", { pageTitle: "Video Detail" });
export const editVideo = (req, res) => res.render("editVideo", { pageTitle: "Edit Video" });
export const deleteVideo = (req, res) => res.render("deleteVideo", { pageTitle: "Delete Video" });
```
